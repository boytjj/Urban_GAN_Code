# Compare sample block images & building images to eliminate outliers

```python
from PIL import Image
from PIL import ImageColor
import numpy as np
import glob
import os
```

```python
blk_list = ['new_blk1', 'new_blk2', 'new_blk3', 'new_blk4', 'new_blk5' ]
bld_list = ['new_bld1', 'new_bld2', 'new_bld3', 'new_bld4', 'new_bld5' ]
```

```python
# Obtain remaining file numbers in blks
def remain_blks(path):
    '''
    Obtain remaining block files from block folder
    Input: str / Output: List
    '''
    file_list = glob.glob(path+'/*.png')
    name_list = list(map(lambda x: x.split('\\')[1], file_list))
    return name_list

blk_names = remain_blks(blk_list[0])
```

```python
# Obtain building file names
def get_bldnames(name_list):
    '''
    Obtain name list of remaining buildings
    Input: List / Output: List
    '''
    bld_list = list(map(lambda x: 'sample_building_' + x.split('_')[1], name_list))
    return bld_list

bld_names = get_bldnames(blk_names)
print(bld_names[:10])
```

```python
# Remain file if name is correct
def get_blds(remain_list, path, i):
    '''
    Remain building files accords with name lists
    Input: List, str, int / Output: folder
    '''
    for filename in os.listdir(path):
        if filename in remain_list:
            print(filename)
            img = Image.open(path+'/'+ filename)
            img.save('./remain_blds{}/remain_'.format(i) + filename)
        
    return

get_blds(bld_names[:10], 'new_bld1', 1)
```

```python
# Merge functions
def compare_blkbld(blk, bld, i):
    '''
    Remain blds accords with blks
    Input: str, str, integer / Output: folder
    '''
    target_blds = remain_blks(blk)
    target_bldnames = get_bldnames(target_blds)
    save_blds = get_blds(target_bldnames, bld, i)
    
    return

#compare_blkbld('new_blk1', 'new_bld1', 1)
```
